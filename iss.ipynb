```python
import requests
import time
import matplotlib.pyplot as plt

url = "http://api.open-notify.org/iss-now.json"

latitudes = []
longitudes = []
timestamps = []
for i in range(100):
    res = requests.get(url)
    json_data = res.json()
    latitude = float(json_data["iss_position"]["latitude"])
    longitude = float(json_data["iss_position"]["longitude"])
    timestamp = int(json_data["timestamp"])
    
    latitudes.append(latitude)
    longitudes.append(longitude)
    timestamps.append(timestamp)
    
    time.sleep(10)

plt.scatter(longitudes, latitudes, c=timestamps, cmap='viridis', marker='o')
plt.colorbar(label='Timestamp')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of the ISS')

plt.grid(True)
plt.show()
```
